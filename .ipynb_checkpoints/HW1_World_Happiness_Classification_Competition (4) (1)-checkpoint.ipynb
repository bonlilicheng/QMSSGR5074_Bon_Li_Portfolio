```python
# Load data
import pandas as pd
X_train = pd.read_csv('world_happiness_competition_data/X_train.csv')
X_test = pd.read_csv('world_happiness_competition_data/X_test.csv')
y_train = pd.read_csv('world_happiness_competition_data/y_train.csv')
y_train_labels = y_train.idxmax(axis=1)

X_train.head()
```

##2.   Preprocess data using Sklearn Column Transformer/ Write and Save Preprocessor function


```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# We create the preprocessing pipelines for both numeric and categorical data.

numeric_features = X_train.drop(['Country or region', 'name', 'region', 'sub-region'], axis=1)
numeric_features=numeric_features.columns.tolist()

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

categorical_features = ['region', 'sub-region']

#Replacing missing values with Modal value and then one hot encoding.
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# final preprocessor object set up with ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

#Fit your preprocessor object
preprocess=preprocessor.fit(X_train) 
```

```python
# Write function to transform data with preprocessor

def preprocessor(data):
    data.drop(['Country or region', 'name'], axis=1)
    preprocessed_data=preprocess.transform(data)
    return preprocessed_data
```

```python
# check shape of X data after preprocessing it using our new function
preprocessor(X_train).shape
```

## Examine features that predict happiness categories using one or more models that allow for automatic feature selection

### SelectPercentile feature selection

In the feature selection model, I chose SelectPercentile as my first feature selection model.SelectPercentile is a feature selection method that selects a certain percentage of the best features based on their statistical scores. It is often used in machine learning to improve model performance by reducing the number of features, which can reduce overfitting, increase interpretability, and improve computational efficiency. The percentile of features to select is specified as a parameter, allowing for a trade-off between the number of features and the performance of the model. By selecting only the most relevant features, SelectPercentile can help improve the accuracy and generalization of the model.

I used random forest classier, MLP classifier, decision tree,gradient boost classifier. On the data after SelectPercentile feature selection. You can see the result on the legendboard.

```python
from sklearn.feature_selection import SelectPercentile
select = SelectPercentile(percentile=50)
select.fit(preprocessor(X_train), y_train_labels)
X_train_selected = select.transform(preprocessor(X_train))
X_test_selected = select.transform(preprocessor(X_test))

print('shape before SelectPercentile：{}'.format(preprocessor(X_train).shape))
print('shape after SelectPercentile:{}'.format(X_train_selected.shape))
print('shape after SelectPercentile:{}'.format(X_test_selected.shape))
```

```python
mask = select.get_support()
print(mask)
```

```python
plt.matshow(mask.reshape(1,-1),cmap=plt.cm.cool)
plt.xlabel("Features Selected")
plt.show()
```


## RFE Feature Selection

I used RFE Feature Selection as my second method of Feature selection. RFE is a feature selection method that helps to identify the most important features in a dataset by recursively eliminating the least important ones. It evaluates the performance of a model trained on the data to determine the importance of each feature. By removing the least important features, RFE can improve the accuracy and efficiency of the model. It also helps reduce the risk of overfitting and increase interpretability by reducing the number of features.

After RFE feature selection, I used GradientBoostingClassifier and tuned parameters in GradientBoostingClassifier. You can also see the result in the legend board.

```python
from sklearn.feature_selection import RFE
rfe = RFE(RandomForestClassifier(n_estimators=100,random_state=38),n_features_to_select=12)
rfe.fit(preprocessor(X_train), y_train_labels)
mask = rfe.get_support()
print(mask)
```

```python
plt.matshow(mask.reshape(1,-1), cmap=plt.cm.cool)
plt.xlabel('Features Selected')
plt.show()
```

```python
X_train_rfe = rfe.transform(preprocessor(X_train))
X_test_rfe = rfe.transform(preprocessor(X_test))
```

```python
gbc_rfe = GradientBoostingClassifier(n_estimators=150, learning_rate=0.1,
    max_depth=5, random_state=42)
gbc_rfe.fit(X_train_rfe, y_train_labels)
print('GradientBoostingClassifier score after RFE selection of features{:.5f}'.format(gbc_rfe.score(X_train_rfe, y_train_labels)))
```
